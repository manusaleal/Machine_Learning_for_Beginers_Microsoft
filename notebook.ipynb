Começaremos com o conjunto de dados (ou dataset) padrão do scikit-learn sobre diabetes. Ele possui 442 linhas com informações de pacientes diabéticos e 10 colunas, sendo elas:

1. Idade (anos)
2. Sexo
3. IMC
4. Pressão sanguínea média
5. s1 tc, colesterol sérico total
6. s2 ldl, lipoproteínas de baixa densidade
7. s3 hdl, lipoproteínas de alta densidade
8. s4 tch, colesterol total / HDL
9. s5 ltg, log possível do nível sérico de triglicerídeos
10. s6 glu, nível de açúcar sanguíneo

Vamos carregá-lo para então visualizá-lo:

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection

#Carregando o dataset
X,y=datasets.load_diabetes(return_X_y=True)

#Verificando o dataset
print(X.shape) 
print(X[0]) 
```

return_X_y=True retorna o formato Tupla com a variável independente X e a variável dependente y, sendo esta a medida da doença em questão. O "True" retorna o numpay arrays de modo que o scikit learn já esteja normalizando-os de forma automática (normalização utilizada: as variáveis de recurso são centradas na média e dimensionadas pelo desvio padrão vezes a raiz quadrada de n_samples=442)

Obs.: Os dados já começam na primeira linha. Não há cabeçalho.

```python
#Pegando somente os dados sobre bmi body mass index
X=X[:,2]
print(X.shape)

#Transformando de ndarray 1D (442,)-> 2D (442,1)
X=X[:,None]
```

Conforme aprendemos no curso de introdução a ML do Kaggle e também neste curso, temos que obter os dados de treino e de teste para conseguirmos treinar e depois validar o nosso modelo:

```python
#Dividindo em dados de treino e teste
X_treino, X_teste, y_treino, y_teste=model_selection.train_test_split(X,y, test_size=0.33)
```

Utilizamos 2/3 para treino e 1/3 para teste (por isso 0,33). Agora que temos os nossos dados de treino e de teste, podemos criar o nosso modelo:

```python
#Criando o modelo de regressão linear
modelo=linear_model.LinearRegression()

#Ajustando o modelo
modelo.fit(X_treino, y_treino)
```

Criado e ajustado o modelo com os dados de teste, agora podemos predizer através das características de validação:

```python
#Predição do modelo
y_predict=modelo.predict(X_teste)
```

Vamos plotar nosso gráfico com a predição em relação aos dados de teste:

```python
#Plotando o gráfico 
plt.scatter(X_teste,y_teste, color="black") #dados de teste
plt.plot(X_teste,y_predict, color="blue", linewidth=3) #ppredição

plt.xlabel('IMC')
plt.ylabel('Progressão da diabetes')
plt.title('Progressão da diabetes vs IMC')
plt.show()
```

Com isso, podemos visualizar que quanto maior o IMC, maior a progressão da diabetes.
